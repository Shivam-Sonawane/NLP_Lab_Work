```python
d = train_dataset[:10]['article']
print(d)
```

```python
# Convert the list to a string
text_string = d[0]

# Print the resulting string
print(text_string)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text_string)
```

```python
doc
```

```python
for np in doc.noun_chunks:
  print(np)
```

```python
for token in doc:
  print(token.text,"====>>>>", token.pos_)
```

```python
noun1 = ([token.lemma_ for token in doc if token.pos_ == "NOUN"])
print(noun1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec = TfidfVectorizer()
tfidf_matrix = vec.fit_transform(noun1)
```

```python
context_matrix = tfidf_matrix.toarray()
print("NN===>>>", noun1)
```

```python
print(context_matrix)
```

```python
context_matrix.shape
```

```python
context_matrix.size
```

```python
terms = vec.get_feature_names_out()
print(terms)
```

```python
noun_phrase = [chunk.text for chunk in doc.noun_chunks]
print("Noun_phrase:-", noun_phrase)
```

```python

for i in range(len(noun_phrase) - 1):
    token1 = nlp(noun_phrase[i])
    token2 = nlp(noun_phrase[i + 1])
    similarity = token1.similarity(token2)
    sim_value= (f"Similarity between '({noun_phrase[i]}' and '{noun_phrase[i + 1]})': {similarity}")
    print(sim_value)
```

```python
import pandas as pd
similarity_data=  []
similarity_data.append(sim_value)
```

```python
# Create a DataFrame
df_similarity = pd.DataFrame(similarity_data, columns=["Noun Phrase 1", "Noun Phrase 2", "Similarity"])

from tabulate import tabulate
print(tabulate(df_similarity, headers="keys", tablefmt="pretty", showindex=False))


# Print the DataFrame
print(df_similarity)
```
